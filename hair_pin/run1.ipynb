```python
# Results Comparison and Visualization

import pickle
import matplotlib.pyplot as plt
import numpy as np

with open('./save/loss_acc_ratio_0.0.pkl', 'rb') as f:
    data_ratio0 = pickle.load(f)

with open('./save/loss_acc_ratio_0.25.pkl', 'rb') as f:
    data_ratio025 = pickle.load(f)

fig, ax = plt.subplots(figsize=(12, 6))
local_ep = data_ratio0.get('local_ep', 5) 
total_local_epochs_0 = range(1, len(data_ratio0['client_A_r2']) + 1)
total_local_epochs_025 = range(1, len(data_ratio025['client_A_r2']) + 1)

# Pure FL (ratio=0)
ax.plot(total_local_epochs_0, data_ratio0['client_A_r2'], '^-', 
        label='Designer A for 3rd order HPF in pure FL', 
        linewidth=2, markersize=4, color='orange', alpha=0.8)

ax.plot(total_local_epochs_0, data_ratio0['client_B_r2'], 'v-', 
        label='Designer B for 7th order HPF in pure FL', 
        linewidth=2, markersize=4, color='purple', alpha=0.8)

# Global validation R2
val_epochs_0 = [(i+1) * local_ep for i in range(len(data_ratio0['val_r2']))]
ax.plot(val_epochs_0, data_ratio0['val_r2'], 'D-', 
        label='Global model for 5th order HPF in pure FL', 
        linewidth=2.5, markersize=7, color='black', markerfacecolor='black')

# Encrypted FL (ratio=0.25)
ax.plot(total_local_epochs_025, data_ratio025['client_A_r2'], '^-', 
        label='Designer A for 3rd order HPF in encrypted FL (k=0.25)', 
        linewidth=2, markersize=4, color='blue', alpha=0.8)

ax.plot(total_local_epochs_025, data_ratio025['client_B_r2'], 'v-', 
        label='Designer B for 7th order HPF in encrypted FL (k=0.25)', 
        linewidth=2, markersize=4, color='darkorange', alpha=0.8)

val_epochs_025 = [(i+1) * local_ep for i in range(len(data_ratio025['val_r2']))]
ax.plot(val_epochs_025, data_ratio025['val_r2'], 's-', 
        label='Global model for 5th order HPF in encrypted FL (k=0.25)', 
        linewidth=2.5, markersize=7, color='red', markerfacecolor='red')

ax.set_xlabel('Local epoch', fontsize=13)
ax.set_ylabel('R2', fontsize=13)
ax.set_ylim(-0.2, 1)
ax.set_title('R2 Comparison: Training Process with All Local Epochs', fontsize=14, fontweight='bold')
ax.legend(fontsize=9, loc='center right')
ax.grid(True, alpha=0.3)

plt.tight_layout()
plt.show()
```
